```python
def frequency_range(spectrogram,sample_rate,split_frequency):
  freq_range = sample_rate/2
  freq_delta_per_bin = freq_range/spectrogram.shape[0]
  split_freq = np.floor(split_frequency/freq_delta_per_bin)
  return int(split_freq)
```

```python
frequency_range(SPEC,sr7_1,81950)
```

```python
def _BER_(spectrogram,sample_rate,split_frequency):
  split_freq_bin = frequency_range(spectrogram,sample_rate,split_frequency)

  power_spec = (np.abs(spectrogram)**2)
  BER = []

  for i in power_spec:
    low_freq = np.sum(i[:split_freq_bin])
    high_freq = np.sum(i[split_freq_bin:])
    ber_current = low_freq/high_freq

    BER.append(ber_current)
  print(power_spec.shape)
  return np.array(BER)
```

```python
BER1 = _BER_(SPEC,sr7_1,2000)

```

```python
BER2 = _BER_(SPEC2,sr9_1,2000)

```

```python
time = librosa.frames_to_time(range(len(BER1)),hop_length=2)


plt.plot(time, BER1, label = "first audio signal")
plt.plot(time, BER2, label = "second audio signal")
plt.legend()
plt.ylim((0, 3))
plt.show()
```
