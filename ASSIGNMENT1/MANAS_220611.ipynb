LABEL ENCODING THE COMPANY NAMES FOR FURTHER GRAPHICAL ANALYSIS

```python
from sklearn.preprocessing import LabelEncoder # OneHotEncoding, Scaling & normalization, Splitting test/train data

le = LabelEncoder()
df['Company_Encoded'] = le.fit_transform(df['Company'])

# Create a mapping dictionary
encoding_mapping = dict(zip(df['Company_Encoded'].unique(), df['Company'].unique()))

# Display the mapping
print("Mapping between Original Values and Encoded Values:")
for original_value, encoded_value in encoding_mapping.items():
    print(f"{original_value}: {encoded_value}")
```

```python
df['TypeName'] = df['TypeName'].str.replace(' ', '_')
df['TypeName'] = df['TypeName'].str.replace('2_in_1', 'two_in_one')
```

HERE I WILL DO THE ONE HOT ENCODING FOR 'TYPENAME' COLUMN AS I FEEL THAT IF SOMEONE IS LOOKING FOR GAMING LAPTOP HE CAN DIRECTLY SEE THE GAMING COLUMN WHICH WILL BE CREATED AFTER ONE HOT ENCODING

```python
df['TypeName'].unique()
```

```python
df['TypeName']
```

```python
df.columns
```

```python
df = pd.get_dummies(df,columns =['TypeName'],drop_first = True)
```

```python
df.columns
```

```python
df.columns = df.columns.to_series().apply(lambda x: x.replace('TypeName_', ''))
```

```python
df.columns
```

```python
df['Ram'] = df['Ram'].apply(lambda x: x.split('G')[0])
df['Ram'] = pd.Series(df['Ram']).astype(int)
```

```python
df.rename(columns = {'Ram':'RAM(in GB)'},inplace = True)
```

```python
df
```

```python
df.info()
```

```python

df['Frequency(GHz)'] = df['Cpu'].str.extract(r'(\d+\.\d+)GHz')
df['Cpu'] = df['Cpu'].str.replace(r'\s*\d+\.\d+GHz', '', regex=True)


df
```

```python
df['Cpu'].unique()
```

AGAIN FOR THE SAME REASON AS 'TYPENAME' , I AM ONE-HOT ENCODING THE PROECESSOR COLUMN, LIKE IF SOMEONE WANTS TO BUY INTEL PROCESSOR HE CAN DIRECTLY SEE THE INTEL COLUMN

```python

df['Intel_Processor'] = df['Cpu'].apply(lambda x: x.replace('Intel ', '') if 'Intel' in x else 0).fillna(0)
df['AMD_Processor'] = df['Cpu'].apply(lambda x: x.replace('AMD ', '') if 'AMD' in x else 0).fillna(0)
df['Samsung_Processor'] = df['Cpu'].apply(lambda x: x.replace('Samsung ', '') if 'Samsung' in x else 0).fillna(0)

```

```python
df
```

AGAIN THE SAME REASON FOR ONE-HOT ENCODING

```python
df['Memory'].unique()
```

```python
df['Memory'] = df['Memory'].str.strip()
```

```python
df
```

```python
import re
df['Memory(in_GB)'] = df['Memory'].apply(lambda x: re.findall(r'\d+', x))
df['Memory(in_GB)'] = df.apply(lambda row: int(row['Memory(in_GB)'][0]) * 1000 if 'TB' in row['Memory'] and row['Memory(in_GB)'] else int(row['Memory(in_GB)'][0]) if row['Memory(in_GB)'] else 0, axis=1)


df['SSD(in_GB)'] = df.apply(lambda row: row['Memory(in_GB)'] if 'SSD' in row['Memory'] else 0, axis=1)
df['HDD(in_GB)'] = df.apply(lambda row: row['Memory(in_GB)'] if 'HDD' in row['Memory'] else 0, axis=1)
df['HYBRID(in_GB)'] = df.apply(lambda row: row['Memory(in_GB)'] if 'Hybrid' in row['Memory'] else 0, axis=1)
df['Flash_Storage(in_GB)'] = df.apply(lambda row: row['Memory(in_GB)'] if 'Flash Storage' in row['Memory'] else 0, axis=1)
```

DOING THIS LABEL ANALYSIS FOR THE PURPOSE OF GRAPHICAL ANALYSIS

```python
le = LabelEncoder()
df['Intel_Processor_encoded'] = le.fit_transform(df['Intel_Processor'].astype(str))
df['AMD_Processor_encoded'] = le.fit_transform(df['AMD_Processor'].astype(str))
df['Samsung_Processor_encoded'] = le.fit_transform(df['Samsung_Processor'].astype(str))

# Create a mapping dictionary
encoding_mapping = dict(zip(df['Intel_Processor_encoded'].unique(), df['Intel_Processor'].unique()))
encoding_mapping2 = dict(zip(df['AMD_Processor_encoded'].unique(), df['AMD_Processor'].unique()))
encoding_mapping3 = dict(zip(df['Samsung_Processor_encoded'].unique(), df['Samsung_Processor'].unique()))

# Display the mapping
print("Mapping between Original Values and Encoded Values:")
for original_value, encoded_value in encoding_mapping.items():
    print(f"{original_value}: {encoded_value}")

for original_value, encoded_value in encoding_mapping2.items():
    print(f"{original_value}: {encoded_value}")

for original_value, encoded_value in encoding_mapping3.items():
    print(f"{original_value}: {encoded_value}")
```

GENERAL RANGE FOR PRICES FOR DIFFERENT TYPES

```python
features = ['Netbook', 'Notebook', 'Ultrabook', 'Workstation',
       'two_in_one_Convertible']
fig, axes = plt.subplots(ncols=len(features), nrows=1, figsize=(20, 4))

min_price = df['Price'].min()
max_price = df['Price'].max()

for i, feature in enumerate(features):

    if (df[feature] != 0).any():

        df_filtered = df[df[feature] != 0]


        sns.regplot(y=feature, x='Price', data=df_filtered, label=feature, scatter_kws={'s': 55}, ax=axes[i])


        axes[i].set_ylabel('')
        axes[i].set_title(f'Scatterplot for Price vs {feature}')


        axes[i].set_xlim(min_price, max_price)


        axes[i].set_yticks(range(0, 3))

    else:
        print(f"No non-zero values for {feature}. Scatter plot not generated.")


axes[-1].set_xlabel('Price')
plt.tight_layout()
plt.show()
```
